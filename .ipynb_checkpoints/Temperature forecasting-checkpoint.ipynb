```python
def plot_periodogram(ts, detrend='linear', ax=None):
    fs = pd.Timedelta("1Y") / pd.Timedelta("1D")
    freqencies, spectrum = periodogram(
        ts,
        fs=fs,
        detrend=detrend,
        window="boxcar",
        scaling='spectrum',
    )
    if ax is None:
        _, ax = plt.subplots()
    ax.step(freqencies, spectrum, color="purple")
    ax.set_xscale("log")
    ax.set_xticks([1, 2, 4, 6, 12, 26, 52, 104])
    ax.set_xticklabels(
        [
            "Annual (1)",
            "Semiannual (2)",
            "Quarterly (4)",
            "Bimonthly (6)",
            "Monthly (12)",
            "Biweekly (26)",
            "Weekly (52)",
            "Semiweekly (104)",
        ],
        rotation=30,
    )
    ax.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))
    ax.set_ylabel("Variance")
    ax.set_title("Periodogram")
    return freqencies, spectrum
```

# Part A: Seasonality modeling

As said, we expect this forcesting exercise to be mostly driven by seasonality. Let's make a moving average plot to see what kind of trend this series has. Since this series has daily observations, let's choose a window of 90 days (one season) to smooth over any short-term changes within the year.

```python
moving_average = nyc_temp.rolling(
    window=90,       # 90-day window
    center=True,      # puts the average at the center of the window
    min_periods=45,  # choose about half the window size
).mean()

ax = nyc_temp.squeeze().plot(style=".", color="0.5", legend = None)
moving_average.squeeze().plot(
    ax=ax, linewidth=3, title="NYC temperature - 90-Day Moving Average", legend=None,
)
```

We will model seasonality using Fourier features. To know which Fourier features to include in our seasonality modeling, we plot the periodogram of the frequency distribution.

Energy spectral density describes how the energy of a signal or a time series is distributed with frequency. Here, the term energy is used in the generalized sense of signal processing.

$$ E = \int_{-\infty}^{\infty}|x(t)|^2 dt = \int_{-\infty}^{\infty}|\hat{x}(t)|^2 dt$$
where $\hat{x}$ is the Fourrier transform of x:

$$\hat{x}(f) = \int_{-\infty}^{\infty} e^{-i2\pi ft}x(t)dt$$

The energy spectral density of x at frequency f is defined as:

$$S_{xx}(f) = |\hat{x}(f)|^2$$

```python
freq,spec=plot_periodogram(nyc_temp.temp_f)
```

```python
split_year = 2020
```

```python
nyc_temp_train = nyc_temp[nyc_temp.index.year<split_year].squeeze()
nyc_temp_test = nyc_temp[nyc_temp.index.year>=split_year].squeeze()
```

```python
fourier = CalendarFourier(freq="A", order=10)  # 10 sin/cos pairs for "A"nnual seasonality

dp = DeterministicProcess(
    index=nyc_temp_train.index,
    constant=True,               # dummy feature for bias (y-intercept)
    order=1,                     # trend (order 1 means linear)
    seasonal=True,               # weekly seasonality (indicators)
    additional_terms=[fourier],  # annual seasonality (fourier)
    drop=True,                   # drop terms to avoid collinearity
)

X = dp.in_sample()  # create features for dates in nyc_temp.index
```

```python
fourier.in_sample(nyc_temp_train.index).iloc[:, :2].plot()
```

```python
fourier.in_sample(nyc_temp_train.index).iloc[:, 2:4].plot()
```

Our data contains:

```python
X.columns
```

We can see we add a trend term, the Fourier periodic functions and the constant term which is to fit the intercept.

```python
model = LinearRegression(fit_intercept=False)
_ = model.fit(X, nyc_temp_train)

y_pred = pd.Series(model.predict(X), index=nyc_temp_train.index)
X_fore = dp.out_of_sample(steps=150)
y_fore = pd.Series(model.predict(X_fore), index=X_fore.index)

ax = nyc_temp_train.plot(color='0.25', style='.', title="NYC temperature - Seasonal Forecast", label = 'True temperature datapoints - in-sample')
ax = y_pred.plot(ax=ax, label="Seasonal temperature")
ax = nyc_temp_test.plot(ax = ax, label = 'True temperature datapoints - out-of-sample', linestyle="None", marker="o", markersize = 3)
ax = y_fore.plot(ax=ax, label="Seasonal Forecast", color='C3')
_ = ax.legend()
```

We can see that our seasonal model is able to capture the main variations in the data. We will now build a Gradient Boosted Regression / Support Vector Regression model to fit the spread between our seasonal forecast and the true temperature. The reason we do this is that linear regression excels at extrapolating trends, but can't learn interactions. GBR / SVR excel at learning interactions, but can't extrapolate trends. Thus we will create a "hybrid" forecaster that combine complementary learning algorithms and let the strengths of one make up for the weakness of the other.

```python
import lightgbm as lgbm
from sklearn.svm import SVR
```

```python
#Train the hybrid model
model1 = LinearRegression(fit_intercept=False)
model1.fit(X, nyc_temp_train)
model1_pred = pd.Series(model1.predict(X), index=nyc_temp_train.index)

res_model_1 = nyc_temp_train - model1_pred
model_2 = SVR(C = 10, kernel = 'rbf')
#lgbm.LGBMRegressor(num_leaves = 4, learning_rate = 0.001, n_estimators = 100)
model_2.fit(X.values, res_model_1)
```

```python
#In sample forecast
in_sample_forecast = model1_pred + pd.Series(model_2.predict(X.values), index=nyc_temp_train.index)
```

```python
X_fore = dp.out_of_sample(steps=150)
out_sample_forecast_model_1= pd.Series(model1.predict(X_fore), index=X_fore.index)
out_sample_forecast_model_2= pd.Series(model_2.predict(X_fore), index=X_fore.index)
res_hybrid = out_sample_forecast_model_1 + out_sample_forecast_model_2
```

```python
ax = nyc_temp_train.plot(color='0.25', style='.', title="NYC temperature - Seasonal Forecast", label = 'True temperature datapoints - in-sample')
ax = in_sample_forecast.plot(ax = ax, label = 'In-sample forecast')
ax = res_hybrid.plot(ax = ax, label = 'Out-of-sample forecast')
ax = nyc_temp_test.plot(ax = ax, label = 'True temperature datapoints - out-of-sample', linestyle="None", marker="o", markersize = 3)
ax.legend()
```

We can see that the model works better in this situation.

We can compare the MSE of the model in-sample and out-of-sample.

In-sample:

```python
hybrid_model_in_sample = ((in_sample_forecast - nyc_temp_train)**2).mean()
```

```python
seasonal_model_in_sample = ((y_pred - nyc_temp_train)**2).mean()
```

```python
print(f'In-sample: Seasonal MSE {seasonal_model_in_sample} vs. hybrid MSE {hybrid_model_in_sample}')
```

```python
np.sqrt(hybrid_model_in_sample)
```

```python
np.sqrt(seasonal_model_in_sample)
```

Out-of_sample:

```python
nyc_temp_test, res_hybrid = nyc_temp_test.align(res_hybrid, join="left")
hybrid_model_out_sample = np.nanmean((res_hybrid - nyc_temp_test)**2)
```

```python
seasonal_model_out_sample = ((y_fore - nyc_temp_test)**2).mean()
```

```python
print(f'Out-of-sample: Seasonal MSE {seasonal_model_out_sample} vs. hybrid MSE {hybrid_model_out_sample}')
```

```python
np.sqrt(seasonal_model_out_sample)
```

```python
np.sqrt(hybrid_model_out_sample)
```

We can see that our hybrid model performs better than the initial model both in-sample and out-of-sample. The SVR performs better than the GBR which tends to overfit the training data. An improvement would be to create a validation set where to fine-tune our GBR which most often performs very well if correctly tuned.

# Part B: Prediction interval for 2020-05-14

We now retrain our model on the full sample to create a one day forecast and for the 1-month ahead forecast. 

```python
data = nyc_temp.squeeze()
fourier = CalendarFourier(freq="A", order=10)  # 10 sin/cos pairs for "A"nnual seasonality

dp = DeterministicProcess(
    index=data.index,
    constant=True,               # dummy feature for bias (y-intercept)
    order=1,                     # trend (order 1 means linear)
    seasonal=True,               # weekly seasonality (indicators)
    additional_terms=[fourier],  # annual seasonality (fourier)
    drop=True,                   # drop terms to avoid collinearity
)

X = dp.in_sample()  # create features for dates in nyc_temp.index

#Train the hybrid model
model1 = LinearRegression(fit_intercept=False)
model1.fit(X, data)
model1_pred = pd.Series(model1.predict(X), index=data.index)

res_model_1 = data - model1_pred
model_2 = SVR(C = 10, kernel = 'rbf')
#lgbm.LGBMRegressor(num_leaves = 4, learning_rate = 0.001, n_estimators = 100)
model_2.fit(X.values, res_model_1)
```

```python
in_sample_pred = pd.Series(model1.predict(X), index=X.index) + pd.Series(model_2.predict(X), index=X.index)
```

```python
X_fore = dp.out_of_sample(steps=1)
out_sample_forecast_model_1= pd.Series(model1.predict(X_fore), index=X_fore.index)
out_sample_forecast_model_2= pd.Series(model_2.predict(X_fore), index=X_fore.index)
res_hybrid = out_sample_forecast_model_1 + out_sample_forecast_model_2
```

```python
res_hybrid
```

```python
n = len(nyc_temp)
```

```python
pred = res_hybrid.iloc[-1]
```

```python
pred
```

```python
low = pred - t.ppf(0.95, df = len(X))*np.sqrt(((in_sample_pred - data)**2).mean()*(1 + 1/n + ((pred-data.mean())**2)/((data - data.mean())**2).sum()))
high = pred + t.ppf(0.95, df = len(X))*np.sqrt(((in_sample_pred - data)**2).mean()*(1 + 1/n + ((pred-data.mean())**2)/((data - data.mean())**2).sum()))
```

```python
print(f'The prediction interval is: [{low}, {high}]')
```
